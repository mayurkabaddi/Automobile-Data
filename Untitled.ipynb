```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
df.to_csv("shop_info.csv")
```

```python
# de = df['price']
# plt.figure(figsize=(10,5))
# sns.histplot(data = de, kde=True, bins=30, color='teal')
# plt.title("Distribution of Car Prices")
# plt.xlabel("Price")
# plt.ylabel("Frequency")
# plt.savefig("car_price_distribution1.png", dpi=300)

def classify_tier(price):
    if price < 15000:
        return 'Basic'
    elif price < 40000:
        return 'Regular'
    else:
        return 'Premium'

df['tier'] = df['price'].apply(classify_tier)
```

```python

plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='price', hue='tier', kde=True, bins=30,
             palette={'Basic': 'limegreen', 'Regular': 'royalblue', 'Premium': 'crimson'})
plt.title("Distribution of Car Prices by Tier")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig("car_price_distribution_by_price_range.png", dpi=300)
```

```python

filtered_prices = df[(df['price'] > 5000) & (df['price'] < 20000)]
brand_counts = filtered_prices['make'].value_counts().reset_index()
brand_counts.columns = ['make', 'count']
ag = filtered_prices.groupby(['make','drive-wheels'])['price'].size().reset_index(name = 'count')
plt.figure(figsize=(12, 6))
sns.barplot(data=ag, x='make', y='count', palette='magma',hue='drive-wheels')
plt.title("Number of Cars per Brand (₹5K–₹20K Range)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("car_price_distribution2.png", dpi=300)

plt.show()
```

```python
aa = df.groupby('body-style')['price'].size()
plt.figure(figsize=(12,5))
sns.boxplot(data=filtered_prices, x='body-style', y='price', palette='Set3')
plt.title("Price Distribution by Body Style")
plt.savefig("car_price_distribution3.png", dpi=300)




```

```python
ht = df[['symboling', 'normalized-losses','wheel-base', 'length', 'width', 'height', 'curb-weight','engine-size', 'bore', 'stroke','compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg','highway-mpg', 'price']]
ht = ht.apply(pd.to_numeric, errors='coerce')
corr_matrix = ht.corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Shop Features')
plt.savefig("car_price_distribution8.png", dpi=300)
```
